```python
import pandas as pd
import numpy as np

# Assuming you have your dataset loaded into a DataFrame named 'your_dataset'
# your_dataset = ...

# Number of rows in your dataset
num_rows = len(merged_5k_data)

# List of days of the week in the desired order of preference
days_of_week = ['Sunday', 'Saturday', 'Friday', 'Thursday', 'Wednesday', 'Tuesday', 'Monday']

# Calculate the number of visits for each day based on your preference
# Sunday has the highest count (e.g., 30%), Saturday (e.g., 25%), Friday (e.g., 20%), Monday (e.g., 5%)
visit_percentages = [0.3, 0.25, 0.2, 0.05]

# Calculate the number of visits for each day
total_visits = int(num_rows * 0.9)  # 90% of total rows (to ensure enough rows for the top 4 days)
visit_counts = [int(total_visits * percentage) for percentage in visit_percentages]

# Calculate the remaining number of rows for Tuesday, Wednesday, and Thursday
remaining_rows = num_rows - sum(visit_counts)

# Assign the remaining rows uniformly across Tuesday, Wednesday, and Thursday
remaining_counts = remaining_rows // 3
visit_counts.extend([remaining_counts, remaining_counts, remaining_counts])

# Shuffle the visit counts based on the days_of_week order
np.random.shuffle(visit_counts)

# Create a dictionary to map days of the week to visit counts
visit_count_mapping = dict(zip(days_of_week, visit_counts))

# Generate random shuffled Visiting_day values based on the visit counts
visiting_day_values = np.random.choice(days_of_week, size=num_rows, replace=True, p=[visit_count_mapping[day] / num_rows for day in days_of_week])

# Add the 'Visiting_day' column to your dataset
merged_5k_data['Visiting_day'] = visiting_day_values

# Display the first few rows of the updated DataFrame
print(merged_5k_data.head())
```
